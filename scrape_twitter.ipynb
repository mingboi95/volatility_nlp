```python
import tweepy
from tweepy import OAuthHandler

auth = OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)

api = tweepy.API(auth)
```

```python
import pandas as pd

twitter_handles = pd.read_csv("TwitterHandles.csv")
twitter_handles.head()
```

```python
all_tweets = []

for twitter_handle in twitter_handles['Twitter Handle']:
    try: 
        print(f"Getting tweets from {twitter_handle}")

        user_tweets = []

        new_tweets = api.user_timeline(screen_name = twitter_handle, count = 200, exclude_replies = True, include_rts = False)
        
        user_tweets.extend(new_tweets)
        oldest = user_tweets[-1].id - 1

        while len((new_tweets)) > 0:
            print(f"getting tweets before {oldest}")
            
            #all subsiquent requests use the max_id param to prevent duplicates
            new_tweets = api.user_timeline(screen_name = twitter_handle, count = 200, exclude_replies = True, include_rts = False, max_id = oldest)
            
            #save most recent tweets
            user_tweets.extend(new_tweets)
            
            #update the id of the oldest tweet less one
            oldest = user_tweets[-1].id - 1
            
            print(f"...{len(user_tweets)} tweets downloaded so far")

        all_tweets.extend(user_tweets)
    except:
        print(f"Error getting tweets from {twitter_handle}")
```

```python
df = pd.DataFrame(map(lambda x: x._json, all_tweets))
```
